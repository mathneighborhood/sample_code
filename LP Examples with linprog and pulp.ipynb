```python
# Mod 3
#Exercise 3a checking that complementary slackness gives correct solution

#Same as 2a but in equation standard form

import numpy as np
from scipy.optimize import linprog


c = [1800,15000,18000] 

#Negatives because we need to change the direction from greater than to less than
A = np.array([[-0.4, -2,  -1], 
              [-0.5, -4, -1.5],
              [-0.3, -3, -1.25],
               [-1,-9,-3]])
b = np.array([-35, 
              -47, 
              -30,
              -90])

# Bounds for decision variables 
y1_bounds = (0, None)  
y2_bounds = (0, None)  
y3_bounds = (0, None)  

# Solve the linear programming problem
result = linprog(c, 
                 A_ub=A, 
                 b_ub=b, 
                 A_eq = None,
                 b_eq = None,
                 bounds=[y1_bounds, y2_bounds, y3_bounds], 
                 method='highs')

print(result.x)
print(result.fun)

```

```python
# Mod 12 Exercise 3
# Maximize annual usage
import pulp as lp

# Define the problem
prob = lp.LpProblem("Annual_usage", lp.LpMinimize)

# Decision variables
# x_i is the project type
x = {
    'X1': lp.LpVariable('x_1', lowBound=0, cat='Binary'),
    'X2': lp.LpVariable('x_2', lowBound=0, cat='Binary'),
    'X3': lp.LpVariable('x_3', lowBound=0, cat='Binary'),
    'X4': lp.LpVariable('x_4', lowBound=0, cat='Binary'),
    'X5': lp.LpVariable('x_5', lowBound=0, cat='Binary'),
    'X6': lp.LpVariable('x_6', lowBound=0, cat='Binary'),
    'X7': lp.LpVariable('x_7', lowBound=0, cat='Binary'),
    'X8': lp.LpVariable('x_8', lowBound=0, cat='Binary'),
    'U1': lp.LpVariable('U_1', lowBound=0, cat='Integer'),
    'E1': lp.LpVariable('E_1', lowBound=0, cat='Integer'),
    'U2': lp.LpVariable('U_2', lowBound=0, cat='Integer'),
    'E2': lp.LpVariable('E_2', lowBound=0, cat='Integer'),
    'U3': lp.LpVariable('U_3', lowBound=0, cat='Integer'),
    'E3': lp.LpVariable('E_3', lowBound=0, cat='Integer'),
    'U4': lp.LpVariable('U_4', lowBound=0, cat='Integer'),
    'E4': lp.LpVariable('E_4', lowBound=0, cat='Integer'),
    'U5': lp.LpVariable('U_5', lowBound=0, cat='Integer'),
    'E5': lp.LpVariable('E_5', lowBound=0, cat='Integer'),
}

# Usage
usage = {
    'X1': 4.7, 'X2': 12.5, 'X3': 3.2, 'X4': 7.5, 'X5': 41,  'X6': 47,   'X7': 23,  'X8': 16,
    'U1': 0,   'E1': 0,    'U2': 0,   'E2': 0,  'U3': 0,   'E3': 0,    'U4': 0,   'E4': 0,    
    'U5': 0,   'E5': 0
}

# Costs
cost = {
    'X1': 75,  'X2': 180,  'X3': 350, 'X4': 45,  'X5': 120, 'X6': 80,   'X7': 115, 'X8': 210,
    'U1': 0,   'E1': 0,    'U2': 0,   'E2': 0,   'U3': 0,   'E3': 0,    'U4': 0,   'E4': 0,    
    'U5': 0,   'E5': 0
}

# Acreage
acreage = {
    'X1': 7,   'X2': 12,   'X3': 20,  'X4': 6, 'X5': 3,   'X6': 25,   'X7': 5,   'X8': 8,
    'U1': 0,   'E1': 0,    'U2': 0,   'E2': 0, 'U3': 0,   'E3': 0,    'U4': 0,   'E4': 0,    
    'U5': 0,   'E5': 0
}

weights = {#weights
    'X1': 0,  'X2': 0,  'X3': 0,     'X4': 0, 'X5': 0,  'X6': 0,  'X7': 0,     'X8': 0,
    'U1': 6/16,  'E1': 0,  'U2': 3/16,   'E2': 0, 'U3': 0,  'E3': 2/16,'U4': 1/16,   'E4': 0,    
    'U5': 4/16,'E5': 0
}

goal1 = { # at least 1 priority 1
    'X1': 0,   'X2': 0,   'X3': 1,  'X4': 0, 'X5': 0,   'X6':  1,   'X7': 0,   'X8': 0,
    'U1': 1,   'E1': -1,  'U2': 0,   'E2': 0, 'U3': 0,   'E3': 0,    'U4': 0,   'E4': 0,    
    'U5': 0,   'E5': 0
}

goal2 = { # at least 130k people annually
    'X1': 4.7, 'X2': 12.5, 'X3': 3.2, 'X4': 7.5, 'X5': 41,  'X6': 47,   'X7': 23,  'X8': 16,
    'U1': 0,   'E1': 0,    'U2': 1,   'E2': -1,  'U3': 0,   'E3': 0,    'U4': 0,   'E4': 0,    
    'U5': 0,   'E5': 0
}

goal3 = { #sum of priority scores is at most 7
    'X1': 3,   'X2': 2,    'X3': 1,   'X4': 3, 'X5': 2,   'X6':  1,    'X7': 2,   'X8': 3,
    'U1': 0,   'E1': 0,    'U2': 0,   'E2': 0, 'U3': 1,   'E3': -1,    'U4': 0,   'E4': 0,    
    'U5': 0,   'E5': 0
}

goal4 = { # At least 90% of budget is used (at least 495k)
    'X1': 75,  'X2': 180,  'X3': 350, 'X4': 45,  'X5': 120, 'X6': 80,   'X7': 115, 'X8': 210,
    'U1': 0,   'E1': 0,    'U2': 0,   'E2': 0, 'U3': 0,   'E3': 0,    'U4': 1,   'E4': -1,    
    'U5': 0,   'E5': 0}

goal5 = { # at least 5 projects are selected for implementation
    'X1': 1,   'X2': 1,   'X3': 1,  'X4': 1, 'X5': 1,   'X6':  1,   'X7': 1,   'X8': 1,
    'U1': 0,   'E1': 0,    'U2': 0,   'E2': 0, 'U3': 0,   'E3': 0,    'U4': 0,   'E4': 0,    
    'U5': 1,   'E5': -1}


# Objective function
prob += lp.lpSum(weights[i] * x[i] for i in x)

# Constraints
prob += lp.lpSum(x[i] * goal1[i] for i in x) == 1, "goal 1"
prob += lp.lpSum(x[i] * goal2[i] for i in x) == 130, "goal 2"
prob += lp.lpSum(x[i] * goal3[i] for i in x) == 7,  "goal 3"
prob += lp.lpSum(x[i] * goal4[i] for i in x) == 495, "goal 4"
prob += lp.lpSum(x[i] * goal5[i] for i in x) == 5, "goal 5"

prob += lp.lpSum(x[i] * cost[i] for i in x) <=550, "# cost constraint"
prob += lp.lpSum(x[i] * acreage[i] for i in x) <=50, "# acreage constraint"


# Solve the problem
prob.solve()

# Print the solution
for v in prob.variables():
    print(v.name, "=", v.varValue)

print("Total Cost =", lp.value(prob.objective))

status = prob.status
print("Status:", lp.LpStatus[status])
```
